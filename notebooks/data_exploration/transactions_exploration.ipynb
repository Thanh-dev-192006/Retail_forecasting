# 3. Transactions Volume Distribution

```python
from scipy import stats



# Basic statistics
mean_trans = df_transactions['transactions'].mean()
median_trans = df_transactions['transactions'].median()
std_trans = df_transactions['transactions'].std()
min_trans = df_transactions['transactions'].min()
max_trans = df_transactions['transactions'].max()
q25_trans = df_transactions['transactions'].quantile(0.25)
q75_trans = df_transactions['transactions'].quantile(0.75)

# Distribution shape metrics
skewness_trans = stats.skew(df_transactions['transactions'])
cv_trans = (std_trans / mean_trans) * 100

print("Transaction Volume Distribution (Actual Data Only):")
print("=" * 60)
print(f"Mean transactions/day: {mean_trans:,.0f}")
print(f"Median transactions/day: {median_trans:,.0f}")
print(f"Std deviation: {std_trans:,.0f}")
print(f"Min transactions: {min_trans:,}")
print(f"Max transactions: {max_trans:,}")
print(f"25th percentile: {q25_trans:,.0f}")
print(f"75th percentile: {q75_trans:,.0f}")
print(f"\nDistribution Characteristics:")
print(f"Coefficient of Variation: {cv_trans:.1f}%")
print(f"Skewness: {skewness_trans:.2f}")
print(f"Max/Mean ratio: {max_trans/mean_trans:.1f}x")

```

```python
transactions_by_stores = df_transactions.groupby('store_nbr')['transactions'].mean().sort_values(ascending=False)
print('\nTop 10 Stores by Average Daily Transactions:\n', transactions_by_stores.head(10))
```
